```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import plotly.express as px
import numpy as np
```

1️⃣ - Realize a leitura do dataset breast_cancer

```python
cancer = load_breast_cancer()
x_cancer = cancer.data
y_cancer = cancer.target
```

2️⃣ - Transformando os dados em um DataFrame

```python
x_cancer = pd.DataFrame(x_cancer, columns=cancer['feature_names'])
pd.set_option('display.max_columns', None)
x_cancer['target'] = y_cancer
df_cancer = x_cancer
```

```python
# Itera sobre as colunas e verifica a unicidade
for coluna in df_cancer.columns:
    num_unicos = df_cancer[coluna].nunique()
    num_linhas = df_cancer.shape[0]
    if num_unicos == num_linhas:
        print(f"A coluna '{coluna}' possui valores únicos.")
```

```python
# Contagem de valores únicos por coluna
print("Contagem de valores únicos por coluna:")
print(df_cancer.nunique())
print("-" * 50)

# Análise mais detalhada da coluna 'target'
print("Distribuição da variável 'target':")
print(df_cancer['target'].value_counts())
print("-" * 50)

# Verificando a distribuição em porcentagem
print("Distribuição da variável 'target' em porcentagem:")
print(df_cancer['target'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
```

```python
df_cancer.describe() #Resumo estatistico dos dados
```

```python
#-Distribuicao das classes e contando quantas amostras são malignas (0) e benignas (1).
agrupando_target = df_cancer.groupby(['target']).size()
agrupando_target.index = ['Benigno (0)', 'Maligno (1)']
agrupando_target
```

```python
# Análise descritiva por classe
df_cancer.groupby('target').describe()
```

```python
#-Visualize essa distribuição das classes com um gráfico de barras.
fig = px.bar(x=agrupando_target.index, y=agrupando_target.values,
                title='Distribuição de Amostras (Benigno vs Maligno)',
                text_auto='.2s',
                labels={'x': 'Diagnóstico', 'y': 'Número de Amostras'},
                color=agrupando_target.index,
                color_discrete_sequence=['#d62728', '#1f77b4'],
                text=agrupando_target.values,
                height=400)
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)

fig.update_layout(
    plot_bgcolor='rgba(0,0,0,0)', 
    paper_bgcolor='lightgray',  
    title_x=0.5                  
)

fig.show()
```

```python
# Dicionário para MUDAR AS CORES (usa os valores NUMÉRICOS 0 e 1 como chaves)
color_map = {0: 'blue', 1: 'red'} 

# Dicionário para MUDAR OS RÓTULOS (usa os valores NUMÉRICOS 0 e 1 como chaves)
new_labels = {0: 'Benigno', 1: 'Maligno'}


# 3. Criação do Boxplot Aprimorado
fig = px.box(
    df_cancer,
    x=df_cancer['target'].map(new_labels), # Mapeia os valores 0/1 para os rótulos de texto
    y='mean_radius',
    color='target', # Usa a coluna 'target' (0 e 1) para o mapeamento de cores
    labels={
        'x': 'Diagnóstico', # Nome do Eixo X
        'y': 'Raio Médio',  # Nome do Eixo Y
        'color': 'Diagnóstico' # Nome da Legenda de Cores
    },
    title='Distribuição do Raio Médio por Diagnóstico',
    color_discrete_map=color_map, # Aplica o mapeamento de cores
)

# 4. Melhorias Visuais
fig.update_layout(xaxis={'categoryorder': 'array', 'categoryarray': ['Benigno', 'Maligno']})
fig.show()
```

```python
#-Calcule a correlação entre as colunas.
correlacao = x_cancer.corr()
correlacao.head()
```

```python
# 3. Criar o Heatmap de Correlação
fig = px.imshow(
    correlacao,
    text_auto=True,  # Exibe os valores de correlação na célula
    aspect="auto",   # Ajusta o aspecto para melhor visualização
    color_continuous_scale='RdBu_r',  # Usa um esquema de cores divergente (vermelho/azul)
    range_color=[-1, 1],             # Garante que a escala de cores vá de -1 a 1 (correlação)
    title='Matriz de Correlação entre as Features do Dataset Breast Cancer'
)

# 4. Melhorias no Layout
fig.update_layout(
    height=1000,
    width=1200,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    xaxis_nticks=len(correlacao.columns),
    yaxis_nticks=len(correlacao.columns)
)

fig.show()
```
